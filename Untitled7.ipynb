```python
%%writefile app.py
import streamlit as st
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder

# ========================= MODEL LOAD =========================
def create_demo_model():
    model = tf.keras.Sequential([
        tf.keras.Input(shape=(7,)), # PaySim features 7k thiyenne
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dense(32, activation="relu"),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dense(1, activation="sigmoid")
    ])
    return model

# Setup UI
st.set_page_config(page_title="Fraud Detection Demo", layout="wide")
st.title("🛡️ Federated Learning: Fraud Detection Live Demo")
st.write("Real-time prediction using FedBN trained model.")

# Sidebar for Inputs
st.sidebar.header("Transaction Details")
trans_type = st.sidebar.selectbox("Transaction Type", ["CASH_OUT", "TRANSFER", "CASH_IN", "DEBIT", "PAYMENT"])
amount = st.sidebar.number_input("Transaction Amount", min_value=0.0, value=5000.0)
old_orig = st.sidebar.number_input("Sender Old Balance", min_value=0.0, value=10000.0)
new_orig = st.sidebar.number_input("Sender New Balance", min_value=0.0, value=5000.0)
old_dest = st.sidebar.number_input("Receiver Old Balance", min_value=0.0, value=0.0)
new_dest = st.sidebar.number_input("Receiver New Balance", min_value=0.0, value=5000.0)
step = st.sidebar.slider("Step (Time)", 1, 744, 1)

# Map transaction type (Based on your LabelEncoder)
type_map = {"CASH_OUT": 1, "TRANSFER": 4, "CASH_IN": 0, "DEBIT": 2, "PAYMENT": 3}
type_val = type_map[trans_type]

# Prepare Data
input_data = np.array([[step, type_val, amount, old_orig, new_orig, old_dest, new_dest]], dtype="float32")

# Predict Button
if st.button("Analyze Transaction"):
    model = create_demo_model()
    # Mauluin trained weights thiyanwa nam mehi load karanna puluwan
    # model.load_weights('fedbn_weights.h5')

    prediction = model.predict(input_data)[0][0]

    col1, col2 = st.columns(2)

    with col1:
        st.subheader("Prediction Result")
        if prediction > 0.5:
            st.error(f"🚨 FRAUD DETECTED! (Confidence: {prediction*100:.2f}%)")
        else:
            st.success(f"✅ LEGITIMATE TRANSACTION (Confidence: {(1-prediction)*100:.2f}%)")

    with col2:
        st.subheader("Model Insights (FedBN)")
        st.info("Privacy Status: 100% Secure (No Raw Data Shared)")
        st.write(f"Architecture: Federated Learning")
```
