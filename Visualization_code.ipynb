```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_hybrid_methodology_flow():
    fig, ax = plt.subplots(figsize=(16, 10))
    ax.set_xlim(0, 16)
    ax.set_ylim(0, 10)
    ax.axis('off')

    # Define common styles
    waterfall_color = '#A8DADC' # Soft Teal
    agile_color = '#E63946'      # Soft Red
    doc_color = '#F1FAEE'        # Very Light Green
    text_color = '#1D3557'       # Dark Navy
    arrow_color = '#457B9D'      # Medium Blue

    # Helper function for text boxes
    def draw_box(ax, x, y, width, height, text, color, r=0.1, fontsize=11):
        box = patches.FancyBboxPatch((x, y), width, height, boxstyle=f"round,pad={r}",
                                     ec="black", fc=color, lw=1.5)
        ax.add_patch(box)
        ax.text(x + width/2, y + height/2, text, color=text_color, fontsize=fontsize,
                fontweight='bold', ha='center', va='center', wrap=True)

    # Helper function for arrows
    def draw_arrow(ax, x1, y1, x2, y2, text=""):
        ax.annotate(text, xy=(x2, y2), xytext=(x1, y1),
                    arrowprops=dict(facecolor=arrow_color, edgecolor=arrow_color,
                                    shrink=0.05, width=2, headwidth=8),
                    fontsize=9, color=text_color, ha='center', va='bottom')

    # =========================================================================
    # WATERFALL PHASES (Sequential Planning & Documentation)
    # =========================================================================
    ax.text(2, 9.5, "WATERFALL: SYSTEM FOUNDATIONS", fontsize=14, fontweight='bold', color=text_color, ha='center')

    # 1. Problem & Privacy Planning
    draw_box(ax, 0.5, 8, 3, 1, "PHASE 1:\nProblem Definition &\nPrivacy Constraints (GDPR)", waterfall_color)
    draw_box(ax, 4, 8.25, 2, 0.5, "DOC: Data\nResidency Protocol", doc_color, fontsize=9)
    draw_arrow(ax, 3.5, 8.5, 4, 8.5)

    # 2. Data Architecture
    draw_box(ax, 0.5, 6.5, 3, 1, "PHASE 2:\nData Architecture &\nDirichlet Sampling Design", waterfall_color)
    draw_arrow(ax, 2, 8, 2, 7.5)
    draw_box(ax, 4, 6.75, 2, 0.5, "DOC: Non-IID Split\nSpecification (α=0.5)", doc_color, fontsize=9)
    draw_arrow(ax, 3.5, 7, 4, 7)

    # 3. DNN Stack Selection
    draw_box(ax, 0.5, 5, 3, 1, "PHASE 3:\nTechnology Stack\n(TensorFlow, T4 GPU)", waterfall_color)
    draw_arrow(ax, 2, 6.5, 2, 6)
    draw_box(ax, 4, 5.25, 2, 0.5, "DOC: Model Layers\n& BN Strategy", doc_color, fontsize=9)
    draw_arrow(ax, 3.5, 5.5, 4, 5.5)

    # Waterfall to Agile Gateway
    draw_arrow(ax, 2, 5, 2, 4, "Sequential Input")
    draw_arrow(ax, 3.5, 5, 6.5, 3.5, "Validated Preprocessing")


    # =========================================================================
    # AGILE SPRINTS (Iterative Development & Experimentation)
    # =========================================================================
    ax.text(11, 9.5, "AGILE: ITERATIVE DEVELOPMENT (2-Week Sprints)", fontsize=14, fontweight='bold', color=text_color, ha='center')

    # SPRINT CYCLE
    center_x, center_y = 11, 6
    r_cycle = 2.5

    # Cycle visualization (dashed circle)
    circle = patches.Circle((center_x, center_y), r_cycle, linestyle='--', color=agile_color, fill=False, lw=2)
    ax.add_patch(circle)

    # Sprint Nodes
    draw_box(ax, center_x - 1, center_y + r_cycle - 0.5, 2, 1, "SPRINT PLANNING\n(DNN tuning)", agile_color)
    draw_arrow(ax, center_x + 1, center_y + r_cycle, center_x + 2, center_y + 1)

    draw_box(ax, center_x + r_cycle - 1, center_y - 0.5, 2, 1, "LOCAL TRAINING\n(T4 GPU Updates)", agile_color)
    draw_arrow(ax, center_x + r_cycle, center_y - 1, center_x + 1, center_y - r_cycle + 0.5)

    draw_box(ax, center_x - 1, center_y - r_cycle - 0.5, 2, 1, "SERVER AGGREGATION\n(FedAvg vs FedBN)", agile_color)
    draw_arrow(ax, center_x - 2, center_y - r_cycle, center_x - r_cycle, center_y - 1)

    draw_box(ax, center_x - r_cycle - 1, center_y - 0.5, 2, 1, "SPRINT REVIEW\n(Metric Analysis)", agile_color)
    draw_arrow(ax, center_x - r_cycle, center_y + 1, center_x - 1, center_y + r_cycle - 0.5)

    # Iteration Indicators
    ax.text(center_x, center_y, "20 Rounds\nof FL", fontsize=16, fontweight='bold', color=agile_color, ha='center', va='center')
    draw_arrow(ax, center_x + 0.5, center_y + 0.5, center_x + 1.5, center_y + 1.5)


    # =========================================================================
    # FINAL RELEASE (Output & Evaluation)
    # =========================================================================
    draw_box(ax, 14, 1, 1.5, 3, "FINAL RELEASE:\nValidated FedBN Model\n+ Comparative CSVs", '#1D3557', fontsize=10)
    ax.text(14.75, 2.5, "", color='white', fontweight='bold', ha='center') # Placeholder for white text
    draw_arrow(ax, 12, 1, 14, 1.5)
    draw_arrow(ax, 11, 5.5, 14, 3, "Convergence Achieved")

    # Title
    fig.text(0.5, 0.05, "Figure 3.2: Hybrid Agile-Waterfall Development Methodology",
             ha='center', fontsize=14, fontweight='bold', color=text_color)

    plt.tight_layout()
    plt.savefig("hybrid_methodology_flowchart.png", dpi=300)
    plt.show()

# Run the function to generate the plot
plot_hybrid_methodology_flow()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_hybrid_methodology_flow():
    fig, ax = plt.subplots(figsize=(16, 10))
    ax.set_xlim(0, 16)
    ax.set_ylim(0, 10)
    ax.axis('off')

    # Styles
    waterfall_color = '#A8DADC' # Soft Teal
    agile_color = '#E63946'      # Soft Red
    doc_color = '#F1FAEE'        # Light Cream
    text_color = '#1D3557'       # Dark Navy
    arrow_color = '#457B9D'      # Blue

    # Helper function for text boxes
    def draw_box(ax, x, y, width, height, text, color, r=0.1, fontsize=11):
        box = patches.FancyBboxPatch((x, y), width, height, boxstyle=f"round,pad={r}", ec="black", fc=color, lw=1.5)
        ax.add_patch(box)
        ax.text(x + width/2, y + height/2, text, color=text_color, fontsize=fontsize, fontweight='bold', ha='center', va='center', wrap=True)

    # Helper function for arrows
    def draw_arrow(ax, x1, y1, x2, y2, text=""):
        ax.annotate(text, xy=(x2, y2), xytext=(x1, y1), arrowprops=dict(facecolor=arrow_color, edgecolor=arrow_color, shrink=0.05, width=2, headwidth=8), fontsize=9, ha='center', va='bottom')

    # WATERFALL FLOW (Left Side)
    draw_box(ax, 0.5, 8.5, 3, 1, "Phase 1:\nGDPR Planning &\nPrivacy Protocol", waterfall_color)
    draw_box(ax, 4, 8.75, 2, 0.5, "DOC: Data\nResidency Plan", doc_color, fontsize=9)
    draw_arrow(ax, 3.5, 9, 4, 9)

    draw_box(ax, 0.5, 7, 3, 1, "Phase 2:\nDirichlet (α=0.5)\nSplit Mathematics", waterfall_color)
    draw_arrow(ax, 2, 8.5, 2, 8)
    draw_box(ax, 4, 7.25, 2, 0.5, "DOC: Non-IID\nSpecification", doc_color, fontsize=9)
    draw_arrow(ax, 3.5, 7.5, 4, 7.5)

    draw_box(ax, 0.5, 5.5, 3, 1, "Phase 3:\nTechnology Stack\n(TensorFlow, T4 GPU)", waterfall_color)
    draw_arrow(ax, 2, 7, 2, 6.5)
    draw_box(ax, 4, 5.75, 2, 0.5, "DOC: DNN Model\nArchitecture", doc_color, fontsize=9)
    draw_arrow(ax, 3.5, 6, 4, 6)

    # Transition
    draw_arrow(ax, 2, 5.5, 2, 4.5, "Foundation Locked")
    draw_arrow(ax, 3.5, 5.5, 7, 3.5, "Validated\nPreprocessing")


    # AGILE CYCLES (Center-Right)
    center_x, center_y = 11, 5
    r_cycle = 2.5
    circle = patches.Circle((center_x, center_y), r_cycle, linestyle='--', color=agile_color, fill=False, lw=2)
    ax.add_patch(circle)

    # Sprint Nodes
    draw_box(ax, center_x - 1, center_y + r_cycle - 0.5, 2, 1, "Sprint Planning\n(DNN tuning)", agile_color)
    draw_arrow(ax, center_x + 1, center_y + r_cycle, center_x + 2, center_y + 1)

    draw_box(ax, center_x + r_cycle - 1, center_y - 0.5, 2, 1, "Local Training\n(T4 GPU updates)", agile_color)
    draw_arrow(ax, center_x + r_cycle, center_y - 1, center_x + 1, center_y - r_cycle + 0.5)

    draw_box(ax, center_x - 1, center_y - r_cycle - 0.5, 2, 1, "Aggregation\n(FedAvg vs FedBN)", agile_color)
    draw_arrow(ax, center_x - 2, center_y - r_cycle, center_x - r_cycle, center_y - 1)

    draw_box(ax, center_x - r_cycle - 1, center_y - 0.5, 2, 1, "Sprint Review\n(Metric analysis)", agile_color)
    draw_arrow(ax, center_x - r_cycle, center_y + 1, center_x - 1, center_y + r_cycle - 0.5)

    # Output
    draw_box(ax, 14, 1, 1.5, 3, "Output:\nComparative\nCSVs + Model", agile_color)
    draw_arrow(ax, 12, 1, 14, 1.5)
    draw_arrow(ax, 11, 4.5, 14, 3, "Convergence")

    fig.text(0.5, 0.05, "Figure 3.2: Hybrid Agile-Waterfall Development Methodology Flowchart", ha='center', fontsize=14, fontweight='bold', color=text_color)
    plt.tight_layout()
    plt.show()

# Run to generate and visualize Chapter 3's methodology chart
plot_hybrid_methodology_flow()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, ConnectionPatch

fig, ax = plt.subplots(figsize=(14, 8))
ax.set_xlim(0, 14)
ax.set_ylim(0, 8)
ax.axis('off')

# Central Server
ax.add_patch(FancyBboxPatch((6, 6), 2.2, 1.3, boxstyle="round,pad=0.6", facecolor="#4285F4"))
plt.text(7.1, 6.65, "Central Server\n(FedAvg / FedBN Aggregation)", ha='center', color='white', fontsize=12, fontweight='bold')

# Clients (showing 4 out of 12 for clarity, with labels)
client_x = [1.5, 3.8, 9.2, 11.5]
for i, x in enumerate(client_x):
    ax.add_patch(FancyBboxPatch((x, 2.5), 1.9, 1.3, boxstyle="round,pad=0.5", facecolor="#34A853"))
    plt.text(x + 0.95, 3.15, f"Client {i+1}\n(Local DNN + BN)", ha='center', fontsize=10, color='white')

# Arrows from Server to Clients
plt.annotate("Broadcast Global Model", xy=(7.1, 6), xytext=(2.8, 4.8),
             arrowprops=dict(arrowstyle="->", lw=2, color='black'), ha='center')
plt.annotate("Broadcast Global Model", xy=(7.1, 6), xytext=(10.8, 4.8),
             arrowprops=dict(arrowstyle="->", lw=2, color='black'), ha='center')

# Arrows from Clients back to Server
plt.annotate("Send Model Weights Only", xy=(3.4, 3.8), xytext=(6.2, 5.8),
             arrowprops=dict(arrowstyle="->", lw=2, color='black'))
plt.annotate("Send Model Weights Only", xy=(10.2, 3.8), xytext=(7.8, 5.8),
             arrowprops=dict(arrowstyle="->", lw=2, color='black'))

# FedBN Highlight
plt.text(7, 1.2, "FedBN: Batch Normalization layers kept local on each client\n"
                 "FedAvg: All layers averaged by server",
         ha='center', fontsize=11, bbox=dict(facecolor="lightyellow", alpha=0.9))

plt.title("Figure 2.2.3: Federated Learning Architecture for Privacy Preserving Model Training",
          fontsize=14, pad=25)

plt.savefig("fig_2.2.3_federated_architecture.png", dpi=400, bbox_inches='tight')
plt.close()

print("✅ Saved: fig_2.2.3_federated_architecture.png")
```
