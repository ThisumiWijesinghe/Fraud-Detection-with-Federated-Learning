```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Function to add percentage labels
def add_labels(ax):
    for container in ax.containers:
        ax.bar_label(container, fmt='%.2f%%', padding=3)

# ======================
# 1. PaySim Results Bar Chart
# ======================
paysim_data = pd.DataFrame({
    "Method": ["NoFL", "FedAvg", "FedBN"],
    "Accuracy": [94.42, 99.43, 95.18],
    "Precision": [62.29, 98.19, 82.45],
    "Recall": [52.13, 30.61, 54.75],
    "F1": [45.14, 46.48, 57.35]
})

ax = paysim_data.set_index("Method").plot(
    kind="bar", figsize=(10, 6), colormap="viridis"
)

plt.title("PaySim Dataset - Final Average Performance (%)")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=0)

add_labels(ax)

plt.tight_layout()
plt.savefig("paysim_results.png", dpi=300, bbox_inches='tight')
plt.close()

# ======================
# 2. Credit Card Results Bar Chart
# ======================
credit_data = pd.DataFrame({
    "Method": ["NoFL", "FedAvg", "FedBN"],
    "Accuracy": [73.34, 98.00, 93.00],
    "Precision": [0.14, 5.00, 8.00],
    "Recall": [22.45, 1.00, 12.00],
    "F1": [0.18, 2.00, 5.00]
})

ax = credit_data.set_index("Method").plot(
    kind="bar", figsize=(10, 6), colormap="plasma"
)

plt.title("Credit Card Dataset - Final Average Performance (%)")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=0)

add_labels(ax)

plt.tight_layout()
plt.savefig("credit_results.png", dpi=300, bbox_inches='tight')
plt.close()

# ======================
# 3. Feature Comparison Table (unchanged)
# ======================
fig, ax = plt.subplots(figsize=(12, 5))
ax.axis('off')

table_data = [
    ["Aspect", "Centralized ML", "FedAvg", "My FL + FedBN"],
    ["Raw Data Sharing", "Yes", "No", "No"],
    ["Non-IID Handling", "Poor", "Average", "Good"],
    ["Privacy Level", "Low", "Medium", "High"],
    ["Handles Feature Shift", "No", "Limited", "Yes (FedBN)"],
    ["Evaluation Metrics", "Limited", "Basic", "Full (Acc, Prec, Rec, F1)"]
]

table = ax.table(cellText=table_data, loc='center', cellLoc='center',
                 colWidths=[0.25, 0.25, 0.25, 0.25])

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.8)

plt.title("Feature Comparison Matrix", pad=20)
plt.savefig("feature_comparison_table.png", dpi=300, bbox_inches='tight')
plt.close()

print("Visuals saved with percentage labels!")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 4))
ax.axis('off')

# Step labels
steps = [
    "12 Bank Clients",
    "Non-IID Dirichlet Split",
    "Models:\nNoFL | FedAvg | FedBN",
    "Federated Training",
    "Performance Evaluation"
]

# X positions
x_positions = [0.1, 0.3, 0.5, 0.7, 0.9]

# Draw boxes
for x, step in zip(x_positions, steps):
    plt.text(x, 0.5, step,
             ha='center', va='center',
             bbox=dict(boxstyle="round,pad=0.4", edgecolor="black"))

# Draw arrows
for i in range(len(x_positions) - 1):
    plt.arrow(x_positions[i] + 0.05, 0.5,
              x_positions[i+1] - x_positions[i] - 0.1, 0,
              head_width=0.03, length_includes_head=True)

plt.title("Project Objectives Pipeline")
plt.savefig("objectives_pipeline.png", dpi=300, bbox_inches='tight')
plt.close()
```
